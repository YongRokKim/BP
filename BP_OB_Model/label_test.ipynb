```python
import matplotlib.pyplot as plt
import cv2

# 이미지 파일 불러오기
image = cv2.imread('../BP_OB_Model/Train/Img_000_0012.jpg')

# OpenCV는 BGR 형식을 사용하지만, matplotlib는 RGB 형식을 사용하므로 변환 필요
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 사각형을 그릴 좌표 설정
x1, y1 = 0,318
x2, y2 = x1+740,y1+556

# 사각형 그리기 (이미지, 시작점 좌표, 끝점 좌표, 색상, 선 두께)
cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)

# 이미지를 matplotlib을 사용하여 출력
plt.imshow(image_rgb)
plt.title('Image with Rectangle')
plt.show()
```
